```python
# Function that groups the Pokemon by Type 1 and looks at the average stats

# Total, HP, Attack, Defense, Sp. Atk, Sp. Def, Speed

def bar_template (var):
    color_palette = sns.color_palette('colorblind')
    avg = round(df.groupby('Type 1')[var].mean()).sort_values(ascending=False).head(10)
    avg.plot(kind='bar', color = color_palette)
    plt.title('Top 10 Pokemon Type 1 with Highest ' + var)
    plt.xlabel('Pokemon Type 1')
    plt.ylabel('Average')
    plt.xticks(rotation=45)
    plt.show()
```

```python
bar_template('Sp. Atk')
```

```python
# Function that selects the top 10 PokÃ©mon based on a specified variable: 
# Total, HP, Attack, Defense, Sp. Atk, Sp. Def, Speed

def poke_10 (var):
    top_10 = df[['Name', var]].sort_values(by=var, ascending=False).head(10).set_index('Name')
    return(top_10)
```
